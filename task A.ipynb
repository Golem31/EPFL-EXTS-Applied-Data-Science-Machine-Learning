```python
# Let's find all the entries (rows) that have have a missing value in one of the columns that we will need later on
# and erase them. According to the list of tasks, the columns that we will need for sure are the following:
# 'created_datetime','last_modified_datetime','countries','ingredients_text','fat_100g','carbohydrates_100g','proteins_100g','nutrition-score-uk_100g'.
df3=df2[(df2['fat_100g'].notnull())&(df2['carbohydrates_100g'].notnull())&(df2['proteins_100g'].notnull())&(df2['countries'].notnull())&(df2['countries_tags'].notnull())&(df2['countries_en'].notnull())&(df2['ingredients_text'].notnull())&(df2['nutrition-score-uk_100g'].notnull())]
df3.isnull().sum()
```

```python
# We can see that the dataframe has those columns that we'll need later on full with values, which was our goal in order to
# properly proceed to the analysis in the next tasks in the list of tasks. Let's now look at all the duplicates.
df3.loc[df3.duplicated(),:]
```

```python
# No duplicate. Cool. Now what else...
df3.isnull().sum()
```

```python
# What I will do now may look a bit radical, but since there was no guidelines whatsoever in the way we should "clean"
# the data, I will just keep the data of which I know I'll need it for the next tasks of the list of tasks.
# Erasing all the columns which I won't need later on. I will also erase some columns that seem redundant for
# the futures steps, like for instance there is no need to have the countries data in 3 different formats.
# I will keep the one that has the countries in English ('countries_en') and erase 'countries' (in different languages)
# and 'countries_tags'. The only column I'm keeping that I won't need later on is 'creator' and 'brands'.
df4=df3.drop(['additives_tags','additives_en','serving_size','url','brands_tags','states','states_tags','states_en','nutrition_grade_fr','nutrition-score-fr_100g','saturated-fat_100g','fiber_100g','sugars_100g','energy_100g','countries','countries_tags','created_t','last_modified_t','additives_n','additives','ingredients_from_palm_oil_n','ingredients_that_may_be_from_palm_oil_n','salt_100g','sodium_100g'],axis=1)
df5=df4[(df4['creator'].notnull())&(df4['created_datetime'].notnull())&(df4['product_name'].notnull())&(df4['brands'].notnull())&(df2['brands_tags'].notnull())&(df2['energy_100g'].notnull())]
# This will btw pop up a red-framed of which I have no idea what it means.
```

```python
df5.info()
```

```python
# Here we have a dataframe that finally has no non-null values while preserving as much data as possible in the
# macronutrients', countries' and ingredients' columns. 2 things remained to be done in my view:
# 1. Normalizing the numeric data doesn't make a lot of sense in my view as all ingredients are all measured on a 100g basis.
# But detecting outliers will nevertheless help us to see if there is incorrect or non-valid data (like for instance if an
# ingredient has a negative value, or a value above 100g, which is impossible).
# 2. Have a closer look a the text columns and setting up a common format for all the values.
filter=((df5['fat_100g']>100)|(df5['fat_100g']<0)|(df5['proteins_100g']>100)|(df5['proteins_100g']<0)|(df5['carbohydrates_100g']>100)|(df5['carbohydrates_100g']<0)|(df5['nutrition-score-uk_100g']<0))
df5.loc[filter,:]
```

```python
# One solution would be to replace all values above 100g with 100g and all negative values with 0.
# The problem is we have no idea where this error stems from and we might create totally incorrect values.
# For the purpose of purity of data, I prefer to remove them - there's not a lof of them anyway.
outliers = df5.loc[filter, :]
outliers.shape
```

```python
df6 = df5.drop(outliers.index, axis=0)
```

```python
# Something else we should filter out of the database are products whose macronutrient total go beyond 100g. This is
# something I discovered while already beginning to work on task B of the list of tasks and looking at some distribution
# charts. Some entries have, when adding 'protein_100g', 'carbohydrates_100g', and 'fat_100g', a total of macronutrients
# of more than 100g, which cannot be. Since we don't know which data of those entries are erroneous, we are just going to
# erase those entries.
filter=((df6['fat_100g']+df6['proteins_100g']+df6['carbohydrates_100g'])>100)
df6.loc[filter,:]
```

```python
# Now regarding the text columns. The first problem is that we have text data in different languages, which we can see
# in the 'product_name' and 'ingredients_text' columns, because the data obviously stems from different countries.
# This isn't a problem for product names, as it's normal to expect that products from different countries have
# different names in different languages. The problem is rather in the ingredients, as the list of tasks will require from
# us to analyze this data later on. We will focus on one language only - the one with the most entries - and erase
# the other ones. I hardly expect that it's required from us to translate all non-english data in english...
dftemp=pd.DataFrame(df7.groupby('countries_en').count())
dftemp.sort_values(by=['code'],ascending=False)
# Might not be the most elegant way to create a whole dataframe just in order to figure out
# what is the country with the most entry - didn't find any other way.
```

```python
# OK, we can see that most data come from the US and will most probably contain text data in English only.
# Of course, UK's and Australia's data is going to be English as well and we could very well keep their data as well,
# but for the purpose of the exercise and since US data is broad enough, we'll keep only US data.
df8=df7.loc[df7['countries_en']=='United States']
df8.shape
```

```python
# Later in the list of tasks, we will be asked to analyze the text data of the 'ingredients_text' column and to find out
# the 5 most common ingredients. Let's have a look at some individual 'ingredients_text' entries in order to observe
# to what kind of format issues we will be confronted with.
```

```python
# Some text data is in capital letters, some have ingredients quantified in %, some have parentheses - square or other,
# some composed words have an underscore instead of a space inside of them, and some data contains words that are superfluous
# because they actually aren't an ingredient (like for intance "Filling:", or "evaporated", or "unbleached".
# In principle, nothing of all this will hinder us to do proper text analysis later on in order to find the 5 most common
# ingredients. Some issues will emerge only when we'll effectively perform the text analysis - like if one of the 5 most
# common ingredient is the word "Filling:", then we're sure this word has to be taken out of the column. Therefore, the
# purpose here is not to start task C of the list of tasks already know. There are nevertheless some steps we can already
# make already now, like normalize all in small caps, erase underscores and maybe numbers and %-signs.

ingredients=df8['ingredients_text']

def transform(y):
    y=y.str.lower()
    y=y.str.replace('_',' ')
    y=y.str.replace('-',' ')
    y=y.str.replace('ñ','n')
    y=y.str.replace('à','a')
    y=y.str.replace('é','e')
    y=y.str.replace('è','è')
    y=y.str.replace('[\d]','')
    y=y.str.replace('%','')
    y=y.str.replace('\=','')
    y=y.str.replace('.','')
    y=y.str.replace(';',',')
    y=y.str.replace(':',',')
    y=y.str.replace('\>',',')
    y=y.str.replace('\<',',')
    y=y.str.replace('\"','')
    y=y.str.replace('\?','')
    y=y.str.replace('\#','')
    y=y.str.replace('\@','')
    y=y.str.replace('\°','')
    y=y.str.replace('\$','')
    y=y.str.replace('\+','')
    y=y.str.replace('\§','')
    y=y.str.replace('\|','')
    y=y.str.replace('\*','')
    y=y.str.replace('\[\]','')
    y=y.str.replace('\(\)','')
    y=y.str.replace('\[','(')
    y=y.str.replace('\]',')')
    y=y.str.replace('\{','(')
    y=y.str.replace('\}',')')
    y=y.str.replace(' \)',')')
    y=y.str.replace('\( ','(')
    y=y.str.replace('\(.\)','')
    y=y.str.replace(' ,',',')
    y=y.str.replace('  ',' ')
    return y

ingredients= transform(ingredients)

df8['ingredients_text']=ingredients

# No idea what the warning message below hints at. We weren't taught anything else than the syntax above.
```
