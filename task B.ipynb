```python
# Among the suggested question in part B of the list of tasks, there is "What are the foods that have the highest ratios of
# each macronutrient?". Well... What is meant by "What are the foods" ?... If it's just a question of ranking the food with
# the highest of each macronutrient, we can do that. But there's no chart for that. As for the other question: "What is the
# macronutrient breakdown across different countries?"... Also, not a very welcome question, since we had to focus on the US
# in order to have text data in English only and didn't seriously think you expected from us to translate every single entry.
# "Can you notice some trends or insights about the data?"... Eh...
```

```python
# Usual import commands while also including matplotlip and seaborn for visualization purpose further down.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('task A.csv', index_col=0)
df
```

```python
# I find it personally interesting to see what is the total amount of macronutrients in each product.
total_100g = df['proteins_100g']+df['fat_100g']+df['carbohydrates_100g']
df=pd.concat([df, total_100g.rename('total_100g')], axis=1)
df.head()
```

```python
# The products that contain the most protein are without surprise gelatin powders and optimized protein powders for
# body builders and those who regulary do their workout. Those that contain the least protein are diverse, but these are
# mostly industrially processed food products, specially dessert products that contain sugar.
df.sort_values(by=['proteins_100g'],ascending=False)
```

```python
# The fattest products are oils, whereas the least fat products are similar than the ones that contain the least proteins.
# sweet products, desserts that contain sugar and carbs in general.
df.sort_values(by=['fat_100g'],ascending=False)
```

```python
# The products with the most carbs are industrially processed and sometimes pre-cooked, often containing plenty of sugar,
# while the ones withe the least carbs usually the most natural and simple ones: water, olives, meat, eggs, shrimps, etc.
df.sort_values(by=['carbohydrates_100g'],ascending=False)
```

```python
df.sort_values(by=['total_100g'],ascending=False)
```

```python
# For the chart analysis, let's keep only the numeric data and also keep out the 'code' column, that is also numeric but
# whose data has merely a labelling function for each product, not informing about its content.
df_num=pd.concat([df['proteins_100g'],df['fat_100g'],df['carbohydrates_100g'],df['total_100g']],axis=1)
df_num.head()
```

```python
# Let's have a look at the macronutrients' distribution in all these foods.
```

```python
sns.distplot(df_num.proteins_100g)
plt.show()
```

```python
sns.distplot(df_num.fat_100g)
plt.show()
```

```python
sns.distplot(df_num.carbohydrates_100g)
plt.show()
```

```python
sns.distplot(df_num['total_100g'])
plt.show()
```

```python
# We can see in the distribution charts above that protein and fat have much more similar distribution charts than
# carbohydrates. Products with a very high content of protein or fat are 'rare' whereas they are very numerous when containing
# little or none of them. Which is why we see their distribution chart with very thin heads but fat tails. Products
# containing carbs are a little bit more equally distributed, but the tail is still fatter than the head.
# Finally, the last chart shows that across all foods, a total macronutrient content of 5%-20% is to be expected.
```

```python
# Please note that the charts above have been done with violin plots and box plots as well. They do not provide any additional
# information than the normal distributions, which is why you'll see only them above. Swarm plots have also been attempted,
# but were aborted after 20 min. of calculation without result.
```

```python
# Now one thing that could be interesting to be looking at is if there is and what kind of a relationship exist between the
# macronutrients. Example:
sns.regplot(x=df['proteins_100g'], y=df['fat_100g'], data=df, scatter_kws={'s':1})
```

```python
# In this chart, we can observe than across all foods of our (purged & cleaned) dataset, there is no clear relationship
# between proteins and fat. The distribution expressed by the linear regression line is rather interesting because it vaguely
# ressembles a y=x relationship, meaning that proteins and fat are somehow equivalently distributed acrosss all foods,
# although we can observe very clearly in the scatter that there exist much more foods with a high content of fat and a lower
# content of proteins than the opposite.
```

```python
# Let's look at those relationships between macronutrients more systematically.
# Watch out, it may take quite a few minutes of calculation.
sns.pairplot(df_num, kind='reg')
```

```python
# 1) The 4 charts along the diagonal are merely the same distribution charts that we saw beforehand for each macronutrient.
# 2) As for the other ones, one thing that appears quite clearly if we look at the charts on the right hand side, we can see
# the relationship of each macronutrient's content with the total macronutrient content. In the distribution we can see
# that while it's not uncommon to see carbs to represent 100% of a food's total macronutrient content, it's a bit sparser
# when looking at the top of the chart showing the content of fat vs. the total, and even sparser when looking at the top of
# the chart representing proteins vs. the total. Food products that are consisting only or almost only of proteins are less
# frequent than those that are consisting only of carbs. This is by the way confirmed by the other charts representing the
# content relationship between proteins vs. fat and between proteins vs. carbs. if we look for instance at the 4 charts on
# the very left hand side.
# 3) As observed further above in the single chart representing the content of proteins vs. the content of fat, there are
# products with a high content of fat and a lower content of proteins than the opposite. Nevertheless, the regression line
# shows that in this relationship, overall fat content distribution vs. overall proteins distribution across all products is
# pretty close to equal.
# 4) The chart showing the content of carbs vs. content of proteins is again a bit different: Although there are, just like
# we saw in fat, more products with a higher content of carbs and a lower content of proteins than the opposite, the overall
# distribution of carbs vs. proteins across all products is totally unbalanced as well: It was quite equivalent in the
# regression line of fat vs. proteins, but here the regression line shows that carbs are overall more present than proteins
# across all foods.
```

```python
# One last thing that I did was grouping the products by 'brand'. I would have liked to do it by country, but I took this
# data out of the original dataset since I decided to focus on US data only because of different existing languages in the
# data. So here we are grouping the entries by 'brands' and look at the average content of a macronutrient - proteins, in
# this example - and then rank them according to this value. We could easily do this with carbs and fat as well.
brands=df.groupby('brands').mean()
brands.sort_values(by=['proteins_100g'], ascending=False)
```
