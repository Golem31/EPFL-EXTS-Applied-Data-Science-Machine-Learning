# E. Build a databse

```python
# Import the usual stuff and also sqlite3
import numpy as np
import pandas as pd
import sqlite3
```

```python
# Import the database as cleaned & narrowed down in task A.
df = pd.read_csv('task A.csv', index_col=0)
df
```

```python
# We are going to follow the list of steps as indicated in task E.
```

```python
# "restrict your data to 1000 entries and 5 columns of your choice"
# Let's take a random sample out of the 'df' database.
df1000=df.sample(n=1000)
```

```python
df1000
```

```python
# And let's reduce it to 5 columns of my choice - the choice is actually important for further steps, notably which column
# is going to be the KEY.
df1000=df1000[['code','created_datetime','fat_100g','carbohydrates_100g','proteins_100g']]
```

```python
df1000.head()
```

```python
# "create a connection to a sqlite3 database"
db = sqlite3.connect("task E.db")
```

```python
# "create one or multiple tables, at least one of the tables should have a PRIMARY KEY"
query1= "CREATE TABLE fat (code INTEGER PRIMARY KEY, fat_100g FLOAT64);"
```

```python
query2= "CREATE TABLE carb (code INTEGER PRIMARY KEY, carbohydrates_100g FLOAT64);"
```

```python
query3= "CREATE TABLE prot (code INTEGER PRIMARY KEY, proteins_100g FLOAT64);"
```

```python
cursor = db.cursor()
```

```python
cursor.execute(query1)
```

```python
cursor.execute(query2)
```

```python
cursor.execute(query3)
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
```

```python
results = cursor.fetchall()
print(results)
```

```python
pd.read_sql_query("SELECT tbl_name FROM sqlite_master;", db)
```

```python
 pd.read_sql_query("SELECT * FROM fat;",db)
```

```python
# Well, we learned "Adding data from a pandas DataFrame", but the example shown in the course adds the total dataframe with
# all columns; there's no example of adding only selected columns from a dataframe into a database. Which is why I create 3
# different new sub-dataframes now in order to add them seperately into the database.
df1001=df1000[['code','fat_100g']]
df1001.head()
```

```python
df1002=df1000[['code','carbohydrates_100g']]
df1002.head()
```

```python
df1003=df1000[['code','proteins_100g']]
df1003.head()
```

```python
# "fill the database with your data"
df1001.to_sql(name='fat', con=db, if_exists='append', index=False)
```

```python
df1002.to_sql(name='carb', con=db, if_exists='append', index=False)
```

```python
df1003.to_sql(name='prot', con=db, if_exists='append', index=False)
```

```python
# "run at least one query to demonstrate that it works correctly"
pd.read_sql_query("SELECT * FROM fat LIMIT 5;",db)
```

```python
pd.read_sql_query("SELECT * FROM carb LIMIT 5;",db)
```

```python
pd.read_sql_query("SELECT * FROM prot LIMIT 5;",db)
```
