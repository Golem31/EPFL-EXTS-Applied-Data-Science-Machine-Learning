```python
# B1. Use the column 'Region' to create a separate DataFrame containing the data points from each of the six regions: North America, Latin America, Western Europe, Eastern Europe, Asia Pacific, Africa.
# me: Set the 'Country' as the new index, will help with in tasks B1. to B4.
df2=df2.sort_values('Region')
df2.set_index('Country',inplace=True)
df2
```

```python
# B2. Compute the mean happiness score for each region and rank the regions from most happy to least happy. 
df2.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False)
```

```python
# B4. Compute the difference between the maximum and minimum happiness score for each region. Which region has the largest range of happiness scores?
# me: Creating a DataFrame for this.
dfDifference=pd.DataFrame(df2.groupby('Region')['Happiness Score'].max()-df2.groupby('Region')['Happiness Score'].min())
dfDifference.columns = ['Differences']
dfDifference['Differences'].sort_values(ascending=False)
# me: It's the 'Asia-Pacific' that shows the largest range of 'Happiness Score'.
```

```python
# B3. Compute the number of countries in each region that have a happiness score above 6.0.
# me: Count how many observations (=number of 'Country') exist in each 'Region' and make it a DataFrame.
dfCountByRegion=pd.DataFrame(df2.groupby('Region')['Region'].count())
dfCountByRegion.columns=['Numbers of countries']
dfCountByRegion
```

```python
# B3. Compute the number of countries in each region that have a happiness score above 6.0.
# me: Creating a double loop that, for each 'Region', counts every 'Happiness Score' above 6.0 and sets up a list with that info.
# me: To do this, the DataFrame created above that counts the number of 'Country' per 'Region' is useful.
above6=[]
total=-1
for i in range(dfCountByRegion.shape[0]):
    above=0
    for ii in range(dfCountByRegion.iloc[i,0]):
        total=total+1
        if df2.iloc[total, 1] > 6:    
            above=above+1
    above6.append(above)
above6
# me: Transforming the list created by the double loop into a DataFrame.
above6=pd.DataFrame(above6,columns=['Above 6'])
above6
# me: Resetting numbers as the index of the DataFrame created earlier that counts the number 'Country' per 'Region'.
# me: So we can concatenate both (it seems that both their indexes have to be the same in order to concatenate both DataFrames).
dfCountByRegion.reset_index(inplace=True)
dfCountByRegion=pd.concat([dfCountByRegion,above6],axis=1)
dfCountByRegion
# me: Setting again the 'Region' as the index so we get a nice DataFrame.
dfCountByRegion.set_index('Region',inplace=True)
dfCountByRegion
```

```python
# G1. Compute the probability that randomly selected country has a Happiness score over 6.0 and is from Western Europe. You will have to use pandas to count the appropriate quantities.
# me: Therefore, the probability of randomly selecting a 'Country' that has a 'Happiness Score' above 6 and is from 'Western Europe'
# me: is 17/151 (=11.26%). Let's check that by simulating the process with a loop.
import random
yo=range(df2.shape[0])
experiments = 10000
counter = 0
for i in range(experiments):
    selected = yo[random.randrange(len(yo))]
    if df2['Happiness Score'][selected]>6 and df2['Region'][selected]=='Western Europe':
        counter += 1
counter / experiments
# me: Repeated executions of this loop result indeed into numbers around 0.1126, which fits the estimation above.
# me: Now, I hope I understood your task in G1. correctly, cause the way it is described is anything but clear...
```

```python
# H1. Define a matrix whose rows correspond to countries and the columns to the regions. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the country in row i is in the region in column j and a 0 otherwise.
# me: Well, the way this matrix will look depends a lot on who the DataFrame is ranked or grouped...
# me: In former tasks, I grouped the DataFrame according to 'Region' before reranking it according to the 'Happiness Rank'.
# me: So let's take the DataFrame ranked according to the 'Happiness Rank',
# me: as for the 'Region', let's order them alphabetically.
# me: Let's first have a look at the shape such a matrix would have:
x=np.arange(1,8)
x=x.reshape(1,7)
y=np.arange(1,152)
y=y.reshape(151,1)
hi=np.dot(y,x)
hi
# me: The highest number is 1057, which corresponds to the product 7x151.
# me: We now alphabetically attribute the numbers of the 1st row to the 'Region', hence 1 is 'Africa', 2 is 'Asia-Pacific',
# me: 3 is 'Eastern Europe', 4 is 'Europe', 5 is 'Latin America', 6 is 'North America', and 7 is 'Western Europe',
# me: and the 'Country' is ordered according to the 'Happiness Rank' in the DataFrame.
```

```python
# H1. Define a matrix whose rows correspond to countries and the columns to the regions. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the country in row i is in the region in column j and a 0 otherwise.
# me: What is needed is a double loop in order to check for each entry if the 'Country' (row) is in the right 'Region' (column).
for i in range(151):
    for j in range(7):
        if j==0 and df2['Region'][i]=='Africa':
            hi[i,j]=1
        elif j==1 and df2['Region'][i]=='Asia-Pacific':
            hi[i,j]=1
        elif j==2 and df2['Region'][i]=='Eastern Europe':
            hi[i,j]=1
        elif j==3 and df2['Region'][i]=='Europe':
            hi[i,j]=1
        elif j==4 and df2['Region'][i]=='Latin America':
            hi[i,j]=1
        elif j==5 and df2['Region'][i]=='North America':
            hi[i,j]=1
        elif j==6 and df2['Region'][i]=='Western Europe':
            hi[i,j]=1
        else:
            hi[i,j]=0
# me: Let's make it a DataFrame.
ho = pd.DataFrame(hi)
ho
```
